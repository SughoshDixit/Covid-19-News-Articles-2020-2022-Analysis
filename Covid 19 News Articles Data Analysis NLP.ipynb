## 1. Identify the named entities mentioned in the articles (such as countries, organizations, people, and diseases).

Performing across all the articles

```python
# load the pre-trained NER model
nlp = spacy.load('en_core_web_sm')

# Create EntityRuler instance
ruler = nlp.add_pipe("entity_ruler")

patterns = [
              {"label": "DISEASE", "pattern": "COVID-19"},
              {"label": "DISEASE", "pattern": "covid19"},
              {"label": "DISEASE", "pattern": "Influenza"},
              {"label": "DISEASE", "pattern": "Covid"}
            ]

# Define pattern for new entity
ruler.add_patterns(patterns)

# extract named entities from each article
entities = []
for text in df_sample['title']:
    doc = nlp(text)
    for ent in doc.ents:
        entities.append((ent.text, ent.label_))

# count the occurrences of each named entity type
counts = {}
for entity, label in entities:
    if label in counts:
        counts[label] += 1
    else:
        counts[label] = 1

# print the results
print(counts)
print(entities)
```

```python
output1 = pd.DataFrame(entities,columns=['Name','Entity'])
output1.sort_values(by=['Entity'], inplace = True)
# Display DataFrame
print("Output1:\n",output1,"\n")
```

## 2. Extract the relationships between entities mentioned in the articles (such as "Person X works for Organization Y").

```python
# import necessary libraries
import pandas as pd
import spacy

# load the pre-trained NLP model
nlp = spacy.load('en_core_web_sm')

# create a function to extract entity relationships from text
def extract_relations(text):
    doc = nlp(text)
    relations = []
    for ent in doc.ents:

        if ent.root.dep_ == 'nsubj':
            subj = ent.root.text
            obj = ''
            for child in ent.root.children:
                if child.dep_ == 'dobj':
                    obj = child.text
                    relations.append((subj, obj))

        elif ent.root.dep_ == 'dobj':
            obj = ent.root.text
            subj = ''
            for child in ent.root.children:
                if child.dep_ == 'nsubj':
                    subj = child.text
                    relations.append((subj, obj))
    return relations

# apply the function to each article and create a new column for the entity relationships
df_sample['entity_relationships'] = df_sample['content'].apply(extract_relations)

# print the first 100 rows of the dataframe with entity relationships
print(df_sample['entity_relationships'].head(100))
```

## 3. Identify the key topics and subtopics discussed in the articles (such as "Vaccine development," "Lockdown measures," or "COVID-19 variants").

```python
# we only need to headlines from the data
data_text = df_sample[:300000][['title']]
data_text['index'] = data_text.index
documents = data_text
'''
Loading Gensim and nltk libraries
'''

import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *
import numpy as np
np.random.seed(400)
```

```python
'''
Write a function to perform the pre processing steps on the entire dataset
'''
stemmer = SnowballStemmer("english")
def lemmatize_stemming(text):
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))

# Tokenize and Lemmatize
def preprocess(text):
    result=[]
    for token in gensim.utils.simple_preprocess(text) :
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return result
```

```python
# Import required libraries
import pandas as pd
import numpy as np
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Load the dataset
#data = pd.read_csv('covid_news_articles_2020_2022.csv')

# Preprocess the text data
stop_words = set(stopwords.words('english'))
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    tokens = nltk.word_tokenize(text)
    tokens = [t for t in tokens if t not in stop_words and len(t) > 3]
    return " ".join(tokens)

df_sample['processed_text'] = df_sample['title'].apply(preprocess_text)

# Vectorize the text data
vectorizer = CountVectorizer(max_features=5000, max_df=0.95, min_df=2)
vectorized_data = vectorizer.fit_transform(df_sample['processed_text'])

# Train the LDA model
lda = LatentDirichletAllocation(n_components=10, max_iter=5, learning_method='online', random_state=0)
lda.fit(vectorized_data)

# Print the top topics and subtopics
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
    print(f"Topic {topic_idx}:")
    top_features_idx = topic.argsort()[:-11:-1]
    for i in top_features_idx:
        print(f"\tSubtopic {feature_names[i]}: {topic[i]}")
```

```python
# import necessary libraries
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
from gensim import corpora, models

# load the dataset
#df = pd.read_csv('covid_articles.csv')

# preprocess the text data
stop_words = set(stopwords.words('english'))
lemma = WordNetLemmatizer()

def preprocess(text):
    # convert to lowercase
    text = text.lower()
    # remove punctuation and numbers
    text = re.sub('[^a-z]', ' ', text)
    # tokenize the text
    tokens = nltk.word_tokenize(text)
    # remove stop words
    tokens = [token for token in tokens if token not in stop_words]
    # lemmatize tokens
    tokens = [lemma.lemmatize(token) for token in tokens]
    return tokens

# create a bag of words corpus
corpus = [preprocess(text) for text in df_sample['title']]
dictionary = corpora.Dictionary(corpus)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in corpus]

# fit an LDA model
lda_model = models.LdaModel(doc_term_matrix, num_topics=10, id2word=dictionary, passes=10)

# print the top 5 topics and their keywords
for i, topic in lda_model.show_topics(num_topics=5, num_words=10, formatted=False):
    print(f"Topic {i}:")
    keywords = [word for word, _ in topic]
    print(keywords)
```

```python
unseen_document = "Daily briefing: Why protests haven‚Äô t lead to a surge in COVID-19"

# Data preprocessing step for the unseen document
bow_vector = dictionary.doc2bow(preprocess(unseen_document))

for index, score in sorted(lda_model[bow_vector], key=lambda tup: -1*tup[1]):
    print("Score: {}\t Topic: {}".format(score, lda_model.print_topic(index, 5)))
```

## 4. Analyze the sentiment of the articles towards these topics (positive, negative, or neutral).

```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Load the dataset
#df = pd.read_csv("path/to/covid-news-articles-2020-2022.csv")

# Preprocess the dataset
df_sample = df_sample.dropna(subset=['title', 'content'])
df_sample = df_sample.drop_duplicates(subset=['title', 'content'])

# Define a function to get the sentiment polarity of a given text
def get_sentiment_polarity(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

# Define a function to assign sentiment labels to each article
def get_sentiment_label(polarity):
    if polarity > 0:
        return "positive"
    elif polarity < 0:
        return "negative"
    else:
        return "neutral"

# Analyze the sentiment of the articles
df_sample['sentiment_polarity'] = df_sample.apply(lambda row: get_sentiment_polarity(row['title'] + ' ' + row['content']), axis=1)
df_sample['sentiment_label'] = df_sample.apply(lambda row: get_sentiment_label(row['sentiment_polarity']), axis=1)

# Visualize the sentiment distribution using a bar chart
sentiment_counts = df_sample['sentiment_label'].value_counts()
plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.title('Sentiment Distribution of COVID-19 News Articles')
plt.xlabel('Sentiment Label')
plt.ylabel('Number of Articles')
plt.show()
```
